#Random Selection of Latitude and Longitude within a Bounding Box

When choosing random latitudes and longitudes, it's important to remember that you can't just naively generate uniform random numbers within your study bounds.  This is because lines of longitude converge at the poles, and the distance between two lines of longitude is shorter near the poles than at the equator.  As a result, a simple uniform generation of random points within a lat/lon bounding box will tend to oversample toward the poles.  While this is especially true very near the poles, the effect will happen at all latitudes.

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

In this section, I create 1000 x,y points between -180 and 180, and -90 and 90.  On this "unprojected" view, the points appear evenly randomly distributed with no particular bias.

```python
n = 1000
# bounds are [minx,miny,maxx,maxy]
bounds = [-180.,-90.,180.,90.];

lon = (bounds[2]-bounds[0]) * np.random.rand(n) + bounds[0]
lat = (bounds[3]-bounds[1]) * np.random.rand(n) + bounds[1]

plt.plot(lon,lat,'.')
```

However, when plotted on an orthographic projection, the clustering at high latitudes becomes apparent.

```python
my_map = Basemap(projection='ortho', lat_0=90, lon_0=-90,
              resolution='l', area_thresh=1000.0)
my_map.drawcoastlines()
x,y = my_map(lon,lat)
my_map.plot(x,y,'r.',markersize=4) 
```

The effect is more easily visualized using more points and a binning technique.  Basemap provides a very nice hexbinning option.  The "log" option performs a log transformation on data, and helps with the visualization.  The "ring" around the earth of apparently higher-density points happens because of the viewpoint of the orthographic projection.

```python
n = 1000000
bounds = [-180.,-90.,180.,90.];
lon = (bounds[2]-bounds[0]) * np.random.rand(n) + bounds[0]
lat = (bounds[3]-bounds[1]) * np.random.rand(n) + bounds[1]


my_map = Basemap(projection='ortho',resolution='l',area_thresh=1000.0,
                 lat_0=90, lon_0=-90)
x,y = my_map(lon,lat)
my_map.hexbin(x,y,bins='log')
```

The clustering of points is clearly visible at the poles.  But don't be fooled!  The effect happens everywhere.  See what happens when we change the bounds to a lower latitude selection:

```python
n = 1000000
bounds = [-120.,20.,-60.,60.];
lon = (bounds[2]-bounds[0]) * np.random.rand(n) + bounds[0]
lat = (bounds[3]-bounds[1]) * np.random.rand(n) + bounds[1]


my_map = Basemap(projection='ortho',resolution='l',area_thresh=1000.0,
                 lat_0=40, lon_0=-90)
my_map.drawcoastlines()
x,y = my_map(lon,lat)
my_map.hexbin(x,y)
```

A better way to do this is to generate random points on the surface of the sphere.  You can effectively generate longitude points as above; however, latitude points need to be generated differently.  For a full description of the issue, see Eric Weisstein's excellent article on <a href="http://mathworld.wolfram.com/SpherePointPicking.html">Sphere Point Picking</a>.  The gist is that the angles (0 to 360 and 0 to 180) can be generated (in this case, in radians) by the following:

```python
n = 1000
u = 2.0 * np.pi * np.random.rand(n)
v = np.arccos(2.0 * np.random.rand(n) -1)

# Convert to longitude and latitude
lon = u * 180.0/np.pi - 180.0
lat = v * 180.0/np.pi - 90.0
```

```python
my_map = Basemap(projection='ortho', lat_0=90, lon_0=-90,
              resolution='l', area_thresh=1000.0)
my_map.drawcoastlines()
x,y = my_map(lon,lat)
my_map.plot(x,y,'r.',markersize=4) 
```

In the above orthographic projection, there is no clustering of points at the poles.  Generating points within boundaries can be accomplished with the following code (adapted from above):

```python
def earth_random(shp=1,bounds=(-180.0,-90.0,180.0,90.0)):

    # Convert bounds to radians
    bounds = (np.pi / 180.0) * np.array(bounds)
    
    if shp!=1:    
        u = np.random.rand(shp)
        v = np.random.rand(shp)
    else:
        u = np.random.rand()
        v = np.random.rand()
     
    
    # Scale u to the requested bounds
    u = (bounds[2] - bounds[0]) * u + bounds[0]
    
    # Since latitude converges, generate a distribution based on the cosine
    # of the angles requested, then take the arccos of that range as given in
    # the mathmetica article
    v_range = np.cos(bounds[[3,1]] + np.pi/2)
    v = np.arccos((v_range[1]-v_range[0]) * v + v_range[0]) - np.pi/2
    
    # Convert to degrees
    lon, lat = u * (180.0 / np.pi), v * (180.0 / np.pi)
    
    return lon,lat
```

Now we can generate random points evenly over the sphere:

```python
n = 100000
bounds = [-120.,20.,-60.,60.];
lon, lat =  earth_random(n,bounds)
```

```python
my_map = Basemap(projection='ortho',resolution='l',area_thresh=1000.0,
                 lat_0=40, lon_0=-90)
my_map.drawcoastlines()
x,y = my_map(lon,lat)
my_map.hexbin(x,y)
```

But be advised, if you plot this in "unprojected" space, you'll see the reverse pattern, with points nearer to the poles apparently too thin!

```python
counts, xedges, yedges, Image = plt.hist2d(lon,lat,bins=100)
```

An equal area projection will show the 2D histogram more accurately.

```python
my_map = Basemap(epsg=5070,resolution='l',area_thresh=1000.0,
                 llcrnrlon=bounds[0],llcrnrlat=bounds[1],urcrnrlon=bounds[2],urcrnrlat=bounds[3])
x,y = my_map(lon,lat)
```

```python
counts, xedges, yedges, Image = plt.hist2d(x,y,bins=100)
```
