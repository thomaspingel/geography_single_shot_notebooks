# Preparing the Regression - Joins in Pandas

Just like in ArcGIS, when we want to put two different datasets together, we need to do a join.  To make our lives easier, we're going to use the FIPS/GEOID field in each dataset as our index, and we're going to make sure they come in as strings/text when we load them.  Read more about this in this <a href="http://data-tutorials.com/zip-codes-in-pandas.html" target="_blank">nice article about using zip codes in Pandas</a>.

```python
# Load the data, but specify that geoid is a string
df_election = pd.read_excel('US_County_Level_Presidential_Results_08-16.v2.xlsx', dtype={'geoid':np.str})

# Use geoid as the index/key/unique identifier
df_election = df_election.set_index('geoid')
```

```python
# Do the same for the County Health Rankings data

df_health_16 = pd.read_excel('CHR_2016.xlsx', dtype={'FIPS':np.str})
df_health_16 = df_health_16.set_index('FIPS')
```

```python
# While we don't HAVE to do this, it's often nice to prepare a simple dataframe with only the information we want to use.  

df = pd.DataFrame()
df['trump_prc'] = df_election['gop_2016'] / df_election['total_2016']
df['trump_vote'] = df['trump_prc'] > .5
df['excessive_drinking'] = df_health_16['Excessive drinking - % Excessive Drinking - 2016']

# The join happens automatically, because we've declared compatible indices.  Otherwise, we'd use the Pandas "join" function.
df.head()
```

```python
# Now we can look at the result as a scatterplot

df.plot.scatter('excessive_drinking','trump_prc')
```

```python
# Use seaborn to easily construct a best-fit line:

sns.regplot(x='excessive_drinking',y='trump_prc',data=df)
```

```python
# Or use statsmodels to run an ordinary least squares regression:

ols_results = smf.ols('trump_prc ~ excessive_drinking', data=df).fit()
print(ols_results.summary())
```
